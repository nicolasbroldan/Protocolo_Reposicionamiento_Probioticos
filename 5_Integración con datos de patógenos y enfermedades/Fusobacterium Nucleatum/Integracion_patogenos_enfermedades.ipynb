```python


# Verificar datos y limpiar si es necesario
DisGeNET_df['Genes'] = DisGeNET_df['Genes'].str.split(';')
GeDiPNet_df['Genes'] = GeDiPNet_df['Genes'].str.split(';')
Orphanet_df['Genes'] = Orphanet_df['Genes'].str.split(';')
Virus_Host_df['Genes'] = Virus_Host_df['Genes'].str.split(';')
Intact_df['Alias(es) interactor A'] = Intact_df['Alias(es) interactor A'].str.split('|')
Intact_df['Alias(es) interactor B'] = Intact_df['Alias(es) interactor B'].str.split('|')
```

```python
# Extraer listas de genes
disgenet_genes = set([gene for sublist in DisGeNET_df['Genes'] for gene in sublist])
gedipnet_genes = set([gene for sublist in GeDiPNet_df['Genes'] for gene in sublist])
orphanet_genes = set([gene for sublist in Orphanet_df['Genes'] for gene in sublist])
virus_host_genes = set([gene for sublist in Virus_Host_df['Genes'] for gene in sublist])

# Genes en común
common_genes = disgenet_genes.intersection(gedipnet_genes, orphanet_genes, virus_host_genes)

print(f"Genes comunes entre todas las bases de datos: {common_genes}")
```

```python
import re

# Función para extraer nombres de genes de un alias
def extract_gene_names(alias_list):
    gene_names = []
    if isinstance(alias_list, list):  # Asegurarse de que alias_list es una lista
        for alias in alias_list:
            # Buscar un nombre de gen que siga el patrón 'uniprotkb:GENE_NAME(gene name)'
            match = re.search(r'uniprotkb:([\w-]+)\(gene name\)', alias)
            if match:
                gene_names.append(match.group(1))
    return ','.join(gene_names)  # Convertir la lista de nombres a una cadena separada por comas

# Asegurarse de que las columnas de alias están en el formato adecuado
Intact_df['Alias(es) interactor A'] = Intact_df['Alias(es) interactor A'].apply(lambda x: eval(x) if isinstance(x, str) else x)
Intact_df['Alias(es) interactor B'] = Intact_df['Alias(es) interactor B'].apply(lambda x: eval(x) if isinstance(x, str) else x)

# Aplicar la función a las columnas de alias
Intact_df['Interactor A Genes'] = Intact_df['Alias(es) interactor A'].apply(extract_gene_names)
Intact_df['Interactor B Genes'] = Intact_df['Alias(es) interactor B'].apply(extract_gene_names)

# Verifica el resultado
print(Intact_df[['Interactor A Genes', 'Interactor B Genes']].head())




```

```python

# Limpiar la columna 'Confidence value(s)' para extraer solo el valor numérico
Intact_df_reducido['Confidence value(s)'] = Intact_df_reducido['Confidence value(s)'].str.extract(r'(\d+\.\d+)', expand=False)

# Limpiar las columnas 'Taxid interactor A' y 'Taxid interactor B' para extraer el contenido del último paréntesis
def extract_last_taxid(text):
    # Extraer el contenido después de 'taxid:652616' si está presente
    if "taxid:652616" in text:
        # Buscar la secuencia taxid:652616 y extraer el contenido entre los paréntesis
        match = re.search(r'taxid:652616\("([^"]+)"\)', text)
        if match:
            return match.group(1)
    
    # Si no se encuentra taxid:652616, realizar la limpieza habitual
    # Primera limpieza para eliminar comillas y paréntesis incompletos
    cleaned_text = re.sub(r'\s*".*$', '', text)
    # Encontrar el último paréntesis válido
    matches = re.findall(r'taxid:[\d-]+\(([^)]+)\)', cleaned_text)
    return matches[-1] if matches else text

Intact_df_reducido['Taxid interactor A'] = Intact_df_reducido['Taxid interactor A'].apply(extract_last_taxid)
Intact_df_reducido['Taxid interactor B'] = Intact_df_reducido['Taxid interactor B'].apply(extract_last_taxid)
# Limpiar la columna 'Host organism(s)' para extraer solo el contenido del último paréntesis
Intact_df_reducido['Host organism(s)'] = Intact_df_reducido['Host organism(s)'].apply(extract_last_taxid)

# Mostrar las primeras filas del DataFrame para verificar
Intact_df_reducido.head()
```

```python
# Función para expandir listas de genes en filas individuales
def expand_genes(df, gene_column):
    df_expanded = df.copy()
    # Convertir la columna 'gene_column' en listas de genes
    df_expanded[gene_column] = df_expanded[gene_column].apply(
        lambda x: [gene.strip() for gene in x.split(';')] if isinstance(x, str) else [x]
    )
    # Expandir las listas en filas individuales
    df_expanded = df_expanded.explode(gene_column).reset_index(drop=True)
    return df_expanded
```

```python
# Leer los archivos y crear los DataFrames
DisGeNET_df = pd.read_csv('DisGeNET_table.txt', delimiter='\t')
GeDiPNet_df = pd.read_csv('GeDiPNet_2023_table.txt', delimiter='\t')
Orphanet_df = pd.read_csv('Orphanet_Augmented_2021_table.txt', delimiter='\t')
Virus_Host_df = pd.read_csv('Virus-Host_PPI_P-HIPSTer_2020_table.txt', delimiter='\t')


# Renombrar la columna 'Term' para cada DataFrame
DisGeNET_df = DisGeNET_df.rename(columns={'Term': 'DisGeNET_Term'})
GeDiPNet_df = GeDiPNet_df.rename(columns={'Term': 'GeDiPNet_Term'})
Orphanet_df = Orphanet_df.rename(columns={'Term': 'Orphanet_Term'})
Virus_Host_df = Virus_Host_df.rename(columns={'Term': 'Virus_Host_Term'})
# Expandir listas de genes en todos los DataFrames
DisGeNET_df = expand_genes(DisGeNET_df, 'Genes')
GeDiPNet_df = expand_genes(GeDiPNet_df, 'Genes')
Orphanet_df = expand_genes(Orphanet_df, 'Genes')
Virus_Host_df = expand_genes(Virus_Host_df, 'Genes')
```

```python
# Ejemplo para optimizar columnas y tipos de datos
def optimize_memory(df):
    # Convertir las columnas de tipo objeto (string) a categorías si es posible
    for col in df.select_dtypes(include=['object']).columns:
        df[col] = df[col].astype('category')
    return df

DisGeNET_df = optimize_memory(DisGeNET_df)
GeDiPNet_df = optimize_memory(GeDiPNet_df)
Orphanet_df = optimize_memory(Orphanet_df)
Virus_Host_df = optimize_memory(Virus_Host_df)
```

```python
# Agrupar, concatenar términos y sus respectivos P-values
DisGeNET_agg = DisGeNET_df.groupby('Genes').apply(
    lambda x: pd.Series({
        'DisGeNET_Term': '; '.join(x['DisGeNET_Term']),
        'DisGeNET_Pvalue_Adjusted': '; '.join(x['Adjusted P-value'].astype(str))
    })
).reset_index()

GeDiPNet_agg = GeDiPNet_df.groupby('Genes').apply(
    lambda x: pd.Series({
        'GeDiPNet_Term': '; '.join(x['GeDiPNet_Term']),
        'GeDiPNet_Pvalue_Adjusted': '; '.join(x['Adjusted P-value'].astype(str))
    })
).reset_index()

Orphanet_agg = Orphanet_df.groupby('Genes').apply(
    lambda x: pd.Series({
        'Orphanet_Term': '; '.join(x['Orphanet_Term']),
        'Orphanet_Pvalue_Adjusted': '; '.join(x['Adjusted P-value'].astype(str))
    })
).reset_index()

Virus_Host_agg = Virus_Host_df.groupby('Genes').apply(
    lambda x: pd.Series({
        'Virus_Host_Term': '; '.join(x['Virus_Host_Term']),
        'Virus_Host_Pvalue_Adjusted': '; '.join(x['Adjusted P-value'].astype(str))
    })
).reset_index()

```

```python
# Diccionario de categorías de enfermedades actualizado
disease_categories = {
    "Cáncer": ["carcinoma", "cancer", "tumor", "neoplasm", "leukemia", "lymphoma", "sarcoma", "melanoma"],
    "Enfermedades Infecciosas": ["infection", "viral", "bacterial", "fungal", "parasitic", "malaria", "HIV", "bacteria", "tuberculosis"],
    "Enfermedades Inmunológicas": ["autoimmune", "lupus", "rheumatoid", "asthma", "allergy", "immunodeficiency"],
    "Enfermedades Neurológicas": ["neurological", "amyotrophic", "parkinson", "alzheimer", "epilepsy", "schizophrenia", "bipolar disorder", "autism"],
    "Enfermedades Cardiovasculares": ["cardiovascular", "heart", "hypertension", "myocardial", "stroke", "angina"],
    "Enfermedades Renales": ["kidney failure", "nephropathy", "glomerulonephritis", "nephrotic syndrome"],
    "Trastornos Metabólicos": ["amyloidosis", "glycogen storage disease", "citrullinemia"],
    "Trastornos Genéticos": ["Ehlers-Danlos syndrome", "Charcot-Marie-Tooth disease", "epidermolysis bullosa"],
    "Trastornos de la Piel": ["dermatitis", "ichthyosis", "erythrokeratodermia"],
    "Enfermedades asociadas a la microbiota": [
        "dysbiosis", "gut microbiota", "intestinal microbiota", "helicobacter pylori", "necrotizing enterocolitis", 
        "inflammatory bowel disease", "celiac disease", "acute diarrhea", "non-alcoholic fatty liver disease", 
        "irritable bowel syndrome", "obesity", "diabetes", "inflammatory intestinal disease", 
        "stress", "anxiety", "digestive system", "poor diet", "fiber deficiency", 
        "sedentarism", "alcohol", "tobacco", "diarrhea", "cramps", "constipation"
    ],
    "Desconocido (NaN)": [],  # Aquí se incluirán enfermedades no clasificadas anteriormente
    "Otras Enfermedades": []  # Categoría para valores NaN o no clasificados
}
```

```python
def classify_disease(disease_terms, categories):
    # Si el término es NaN, retornamos 'other'
    if pd.isna(disease_terms):
        return 'Sin información'
    
    # Dividir los términos por ';' y clasificar cada uno
    terms = disease_terms.split(';')
    categories_found = set()
    for term in terms:
        term = term.strip()  # Limpiar espacios en blanco
        for category, keywords in categories.items():
            if any(keyword.lower() in term.lower() for keyword in keywords):
                categories_found.add(category)
                break  # No necesitamos verificar más categorías para este término
    if not categories_found:
        return 'other (NaN)'
    return '; '.join(categories_found)  # Unir las categorías encontradas en una cadena
```

```python
def classify_diseases(row, categories):
    classified_terms = {}
    for col in row.index:
        if 'Term' in col:  # Solo para columnas que contienen términos de enfermedades
            classified_terms[col + '_Category'] = classify_disease(row[col], categories)
    return pd.Series(classified_terms)

# Aplicar la clasificación al DataFrame
classification_df = final_merged_df.apply(classify_diseases, categories=disease_categories, axis=1)

# Agregar las clasificaciones al DataFrame original
final_merged_df = final_merged_df.join(classification_df)

# Mostrar el DataFrame resultante con las nuevas categorías
final_merged_df.head()
```
